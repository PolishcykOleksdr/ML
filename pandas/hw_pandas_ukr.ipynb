```python
import pandas as pd
```

```python
data = pd.read_csv('adult.data.csv')
data.head()
```

**1. Скільки чоловіків и жінок (ознака *sex*) представлено в цьому наборі даних?**

```python
print(data['sex'].value_counts())
```

**2. Який середній вік (ознака *age*) жінок?**

```python
print(data[data['sex'] == 'Female']['age'].mean())
```

**3. Яка доля громадян Німеччини (ознака *native-country*)?**

```python
print(data['native-country'].value_counts(normalize=True)['Germany'])
```

**4-5. Які средні значення та средньоквадратичні відхилення віку тих, хто отримує більше 50K на рік (ознака *salary*) і тих, хто отримує менше 50K на рік?**

```python
data_above_50 = data[data['salary'] == '>50K']['age']

print('Average above 50K: ', data_above_50.mean())
print('Average deviation above 50K: ', data_above_50.std())

data_below_50 = data[data['salary'] == '<=50K']['age']

print('Average below 50K: ', data_below_50.mean())
print('Average deviation below 50K: ', data_below_50.std())
```

**6. Чи правда, що люди, які отримують більше 50k, мають як мінімум вищу освіту? (ознака *education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* или *Doctorate*)**

```python
high_education = ['Bachelors', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', 'Masters', 'Doctorate']

rich_education = data[data['salary'] == '>50K']['education']

is_true = rich_education.isin(high_education).all()

print(is_true)
```

**7. Порахуйте статистику віку для кожної раси (ознака *race*) і кожної статі. Використайте *groupby* и *describe*. Знайдіть таким сином максимальний вік чоловіків раси *Amer-Indian-Eskimo*.**

```python
race_sex_stats = data.groupby(['race', 'sex'])['age'].describe()
max_age = race_sex_stats.xs(('Amer-Indian-Eskimo', 'Male'))['max']

print(max_age)
# print(race_sex_stats)
```

**8. Серед кого більша доля багатіших (>50K): серед одружених чи не одружених чоловіків (ознака *marital-status*)? За одружених вважаємо тих, у кого *marital-status* починається із *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), всі інші - не одружені.**

```python
all_men = data[data['sex'] == 'Male']
rich_men = all_men['salary'] == '>50K'
is_married = all_men['marital-status'].str.startswith('Married')

result = rich_men.groupby(is_married).mean()

print(result)
```

**9. Яку максимальну кількість годин в тиждень працює людина (ознака *hours-per-week*)? Скільки людей працюють таку кількість годин і який відсоток серед них багатих?**

```python
max_hours = data['hours-per-week'].max()
max_hours_people = data[data['hours-per-week'] == max_hours]
rich_percentage = (max_hours_people['salary'] == '>50K').mean()

max_hours_people = max_hours_people.shape[0]

print(max_hours)
print(max_hours_people)
print(rich_percentage)
```

**10. Порахуйте середню тривалість роботи на тиждень (*hours-per-week*) серед багатих і бідних (*salary*) для кожної країни (*native-country*).**

```python
poor_people = data[data['salary'] == '<=50K']
rich_people = data[data['salary'] == '>50K']

poor_salary_country = rich_people.groupby(['native-country'])['hours-per-week'].mean()
rich_salary_country = rich_people.groupby(['native-country'])['hours-per-week'].mean()

print('Poor')
print(poor_salary_country)
print('------------------------')
print('Rich')
print(rich_salary_country)
```
