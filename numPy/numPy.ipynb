```python
import numpy as np

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

data = np.loadtxt(url, delimiter=',', usecols=(range(0, 4)))

print(data)
```

```python
print(data.ndim)
```

```python
first_col_mean = np.mean(data[:, 0])
first_col_median = np.median(data[:, 0])
first_col_standard = np.std(data[:, 0])

print(first_col_mean)
print(first_col_median)
print(first_col_standard)
```

```python
rand = np.random.default_rng()

randoms = rand.choice(data.size, size=20, replace=False)

data.flat[randoms] = np.nan

print(data)
```

```python
column = data[:, 0]
indexes = np.where(np.isnan(column))[0]

print(indexes)
```

```python
data[np.isnan(data)] = 0

print(data)
```

```python
data_filter = (data[:, 2] > 1.5) & (data[:, 0] < 5.0)
datas = data[data_filter]

print(datas)
```

```python
n_columns = data.shape[1]

mid = n_columns // 2

left, right = np.split(data, [mid], axis=1)

print(left)
print(right)
```

```python
idx_left = np.argsort(left[:, 0])
left_sorted = left[idx_left]

idx_right = np.argsort(-right[:, 0])
right_sorted = right[idx_right]

print(left_sorted)
print(right_sorted)
```

```python
union_data = np.hstack((left_sorted, right_sorted))
print(union_data)
```

```python
uniques, counts = np.unique(union_data, return_counts=True)

for val, n in zip(uniques, counts):
    print(f"{val} - {n} times")
```

```python
max_count_idx = np.argmax(counts)

num = uniques[max_count_idx]

print(num)
```

```python
def func(col):
    mean = np.mean(col)

    result = np.where(col < mean, col * 2, col / 4)

    return result
```

```python
print(data[:, 2])
data_func = func(data[:, 2])
print(data_func)
```
